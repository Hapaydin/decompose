```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from datetime import datetime
startTime = datetime.now()
```

```python
df1 = pd.read_csv('/content/gdrive/My Drive/Colab Notebooks/12188M.csv', delimiter=';')
df=df1['Q']
print (df.head())
print ('Mean:',np.mean(df))
```

```python
from statsmodels.tsa.stattools import adfuller
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.tsa.holtwinters import ExponentialSmoothing, SimpleExpSmoothing, Holt
from pylab import rcParams
rcParams['figure.figsize'] = 8, 6
SP=13
PP=15
from statsmodels.tsa.seasonal import STL
decomposition = STL(df, seasonal=SP, period=PP)
res = decomposition.fit()
fig = res.plot()
plt.show()
FileName = "STL-M"+str(SP)+"-"+str(PP)+"-"+str(datetime.now()).replace(":", ".")
fig.savefig(FileName+'.png', dpi=300)
```

```python
trend = res.trend
seasonal = res.seasonal
residual = res.resid
print (len(trend),len(seasonal),len(residual))
print ('T:',trend.mean(),'\nS:',seasonal.mean(),'\nR:',residual.mean())
```

```python
np.savetxt(FileName+'_trend.csv', trend)
np.savetxt(FileName+'_seasonal.csv', seasonal)
np.savetxt(FileName+'_residual.csv', residual)
```
